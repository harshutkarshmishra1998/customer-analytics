1 - Import Libraries

```python
import numpy as np
import pandas as pd
```

2 - Load Database

```python
df_purchase = pd.read_csv('customer-segmented.csv')
df_purchase
```

3 - Convert Segments Into Dummy Variables

```python
segment_dummies = pd.get_dummies(
    df_purchase['Segment'], prefix='Segment', prefix_sep='_')
segment_dummies
```

```python
df_purchase_predictors = pd.concat(
    [df_purchase, segment_dummies], axis=1)
df_purchase_predictors
```

4 - Save Prepared Data

```python
df_purchase_predictors.to_csv("customer-segmented-prepared.csv", sep=",", index=False)
```
